### 找到最近更新信息的20套房子

```python
lj_data['update_day']=lj_data['update_time'].astype(np.datetime64)
lj_data.sort_values(by=['update_day'], ascending=False).head(20)
```

### 平均看房人数

```python
lj_data['view_num'].mean()
```

```python
def func(s):
    res =  "".join([x  for x in str(s) if str.isdigit(x)])
    if res:
        return res
    else:
        return np.nan
```

### 房龄最小的20套房子的平均看房人数、平均面积...

```python
lj_data['build_year'] = lj_data['build_time'].map(func)
lj_data['build_year'].fillna(0, inplace=True)
lj_data['build_year']=lj_data['build_year'].astype(np.int)
lj_data.info()
```

```python
last20_build=lj_data.sort_values(by=['build_year'], ascending=False).head(20)
last20_build['view_num'].mean()
last20_build
```

```python
def func2(s):
    res =  "".join([x  for x in str(s) if str.isdigit(x)])
    if res:
        return int(res)
    else:
        return 0

last20_build['new_area']=last20_build['area'].map(func2)
last20_build['new_area'].mean()
```
