### Pandas操作题(共七个步骤，每个步骤要求及分数均已标明。总分100分)

#### 1.数据导入并展示, 文件位置位于0.Teacher/data/下
   - 请将文件student-info.csv中的内容读入pandas DataFrame "df_info"，注意该文件的分隔符是分号，并展示前五行（7.5分）   
   - 请将文件student-score.csv中的内容读入pandas DataFrame "df_score"，注意该文件的分隔符是分号，并展示前五行（7.5分）

```python
import pandas as pd
df_info = pd.read_csv('/home/hanxiaoyang/sxy-s4/0.Teacher/data/student-info.csv', sep=';', index_col=['ID'])
df_info.head(5)
```

```python
df_score=pd.read_csv('/home/hanxiaoyang/sxy-s4/0.Teacher/data/student-score.csv', sep=';', index_col='ID')
df_score.head(5)
```

#### 2.把两个DataFrame "df_info" 和 "df_score" 按照学生ID对齐拼接起来，结果存在一个新DataFrame "df" 中 (10分）

```python
df = pd.concat([df_info, df_score], axis=1)
df.head()
```

#### 3.把"df"所有的列展示出来(5分)

```python
df.columns
```

#### 4.统计男生的平均分和女生的平均分(G1, G2, G3)，使用groupby和aggregate来操作(10分)

```python
df_group=df.groupby('sex')
df_group[['G1', 'G2', 'G3']].mean()
```

```python
df_group[['G1', 'G2', 'G3']].agg(['mean'])
```

#### 5.统计不同年龄的学生(G1, G2, G3)的平均分，并作出柱状图展示（10分）

```python
age_group=df.groupby('age')
age_group_mean = age_group[['G1', 'G2', 'G3']].mean()
age_group_mean
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
g1=age_group_mean['G1']
g2=age_group_mean['G2']
g3=age_group_mean['G3']
total_width, n = 0.6, 3  
width = total_width / n
x = age_group_mean.index
x = x - (n-1)*width/2 

plt.bar(x, g1, width=width, label='G1')
plt.bar(x+width, g2, width=width, label='G2')
plt.bar(x+2*width, g3, width=width, label='G3')
plt.legend()
```

#### 6.统计不同学校(school一列)的学生的平均分，并作柱状图展示，要求同一科目的两个学校成绩贴在一起展示(10分）

```python
school_group = df.groupby('school')
school_group_mean = school_group[['G1', 'G2', 'G3']].mean()
school_group_mean
```

```python
gp=school_group_mean.loc['GP']
ms=school_group_mean.loc['MS']

total_width, n = 0.6, 3 
width = total_width / 3
count =g1.count()
x = pd.Series(range(count))

plt.bar(x, gp, width=width, label='GP')
plt.bar(x+width, ms, width=width, label='MS')
plt.xticks(x, ('G1', 'G2', 'G3'))
plt.legend()
```

#### 7.统计一下每个学校三个科目总分的前十名，然后把前十名的同学所有信息输出(40分)

```python
df['total'] = df['G1'] +df['G2'] +df['G3']
school_group_total = df.groupby('school')
GP = school_group_total.get_group('GP')
MS= school_group_total.get_group('MS')
GP.sort_values(by=['total'], ascending=False).head(10)
```

```python
MS.sort_values(by=['total'], ascending=False).head(10)
```
